```python
import pandas as pd
from datetime import datetime, timedelta

# Veriler
start_date = datetime(2023, 7, 1)
days = 10
dates = [start_date + timedelta(days=i) for i in range(days)]
correct_words = [75, 81, 71, 86, 90, 93, 90, 95, 96, 96]
total_words = 100
accuracy = [correct / total_words * 100 for correct in correct_words]

# DataFrame oluşturma
data = {
    'Tarih': dates,
    'Doğru Okunan Kelimeler': correct_words,
    'Toplam Kelime Sayısı': [total_words] * days,
    'Doğruluk Yüzdesi': accuracy
}

df = pd.DataFrame(data)

# Tarih sütununu uygun formatta gösterme
df['Tarih'] = df['Tarih'].dt.strftime('%Y-%m-%d')

df
```
